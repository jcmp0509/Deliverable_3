```python
# Step(1): Importing Libraries
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Step(2): Loading iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

```python
# Step(3): Dropping the "species" column
iris = iris.drop(labels= 'species', axis = 1)
iris.head()
```

```python
# Step(4) Adding columns
iris['sepal_sum'] = iris['sepal_length'] + iris['sepal_width']
iris.head()
```

```python
# Step(5)

mean = list(iris.mean())
std = list(iris.std())
min = list(iris.min())
max = list(iris.max())

summary_dict = {'mean' :mean,
                'std' :std,
                'min' :min,
                'man' :max}

summary_df = pd.DataFrame(data = summary_dict)

summary_df
```
